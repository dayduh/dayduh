Visualizing Cases vs Deaths in select CA counties. 

```r
ggplot(cal.counties, aes(x=county, y=cases)) + 
  geom_bar(stat="identity", width=.5, fill="tomato3") + 
  labs(title="Ordered Bar Chart", 
       subtitle="Cases per County in SoCal", 
       caption="source: NYT") + 
  theme(axis.text.x = element_text(angle=65, vjust=0.6))
```

```r
death.plot = cal.counties %>% 
ggplot(aes(x=date,y=deaths, color= county, group = county))+
  geom_line(size = 2, alpha = .75, aes(color=county)) + facet_grid(vars(county), scales = "free") +
  scale_x_date(date_minor_breaks = "2 day", date_labels = "%b-%Y")+
  scale_color_brewer(palette = "Set1")

  
```

```r
print(death.plot)
```

```r
state.data <- state.data %>%
  group_by(state) %>%
  summarise(total_deaths = sum(deaths)) %>% 
  arrange(desc(total_deaths))
head(state.data)
```

```r
usmap = plot_usmap(data = state.data, values = "total_deaths", labels = TRUE)
```

#Using usmap library to visualize total deaths in the United States.

Each map is utilizing different color schemes that are color blind friendly. 

```r
usmap + scale_fill_distiller(type = "seq", palette = "Reds", direction = 0, name = "Total Deaths")
```

```r
usmap + scale_fill_distiller(type = "seq", palette = "YlOrRd", direction = 0, name = "Total Deaths")
```

```r
usmap + scale_fill_viridis(direction = -1, name = "Total Deaths")
```

```r
# If we change the map to black and white we still see the gradient from low to high death count but any text data within the state boundary is compromised. 
usmap + scale_fill_distiller(type = "seq", palette = "Greys", direction = 0, name = "Total Deaths")
```

Using greyscale doesnt look bad but it does make it a bit more difficult to read the state abbreviation or any info that may be within the map. If you dont have any ink available this would suffice but would not be a first choice. 

This visual is pretty bad for people who are red blind. I pasted a line graph from above into a colorblind checker and this is the result. It is difficult to tell the differnce between the SF and Orange counties and the LA and SD areas are pretty similar too.

The manual fill is much better. There is a continuous gradient that is distinguishable for each segment. Unless the person is monochromatic, this would be readible for those suffering from coloblindness.
